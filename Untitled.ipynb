```python
import requests
from bs4 import BeautifulSoup
from flask import Flask

url = 'http://www.gob.ec/api/v1/instituciones'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
instituciones = []
for institucion in soup.find_all('institucion'):
    instituciones.append(institucion.text)
print(', '.join(instituciones))
```
